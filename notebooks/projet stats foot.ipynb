```python
%%time
for x in range(len(matchs)):
    for key in data_df.keys():
        if key in matchs[x].keys():
            data_df.get(key).append(matchs[x].get(key))
    data_df.get('game_id').append(matchs[x].get('id'))
    data_df.get('new_rules').append('False')
    data_df.get('localteam_coach_id').append(matchs[x].get('coaches').get('localteam_coach_id'))
    data_df.get('visitorteam_coach_id').append(matchs[x].get('coaches').get('visitorteam_coach_id'))
    if matchs[x].get('scores').get('ht_score') != None:
        data_df.get('ht_score_local').append(matchs[x].get('scores').get('ht_score')[0])
        data_df.get('ht_score_vis').append(matchs[x].get('scores').get('ht_score')[2])
        data_df.get('ht_score').append(matchs[x].get('scores').get('ht_score'))
    else:
        data_df.get('ht_score_local').append('TbC')
        data_df.get('ht_score_vis').append('TbC')
        data_df.get('ht_score').append('TbC')
    data_df.get('winner_ht').append('TbC')
    if matchs[x].get('scores').get('ft_score') != None:
        data_df.get('ft_score_local').append(matchs[x].get('scores').get('ft_score')[0])
        data_df.get('ft_score_vis').append(matchs[x].get('scores').get('ft_score')[2])
        data_df.get('ft_score').append(matchs[x].get('scores').get('ft_score'))
    else:
        data_df.get('ft_score_local').append('TbC')
        data_df.get('ft_score_vis').append('TbC')
        data_df.get('ft_score').append('TbC')
    data_df.get('winner_ht_vs_ft').append('TbC')
    try:
        data_df.get('subs_loc').append(matchs[x].get('stats').get('data')[0].get('substitutions'))
        data_df.get('subs_vis').append(matchs[x].get('stats').get('data')[1].get('substitutions'))
    except:
        data_df.get('subs_loc').append('TbC')
        data_df.get('subs_vis').append('TbC')
    for y in range(len(matchs[x].get('events').get('data'))):
        if matchs[x].get('events').get('data')[y].get('type')=='substitution':
            for key in subs.keys():
                if key in matchs[x].get('events').get('data')[y].keys():
                    subs.get(key).append(matchs[x].get('events').get('data')[y].get(key)) 
            subs.get('league_id').append(matchs[x].get('league_id'))
            subs.get('event_id').append(matchs[x].get('events').get('data')[y].get('id'))
            subs.get('game_id').append(matchs[x].get('events').get('data')[y].get('fixture_id'))
            subs.get('player_in').append(matchs[x].get('events').get('data')[y].get('player_id'))
            subs.get('player_out').append(matchs[x].get('events').get('data')[y].get('related_player_id'))
            subs.get('sub_number').append('TbC')
            subs.get('subs_goals').append('TbC')
            subs.get('positive_impact').append('TbC')
            subs.get('new_rules').append('False')
        else:
            for key in events.keys():
                if key in matchs[x].get('events').get('data')[y].keys():
                    events.get(key).append(matchs[x].get('events').get('data')[y].get(key))
            events.get('league_id').append(matchs[x].get('league_id'))
            events.get('event_id').append(matchs[x].get('events').get('data')[y].get('id'))
            events.get('event_type').append(matchs[x].get('events').get('data')[y].get('type'))
            events.get('game_id').append(matchs[x].get('events').get('data')[y].get('fixture_id'))
```
